```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('job_offers_20250608_1907.csv')

# Remove empty rows
df_clean = df.dropna(how='all')

# Remove duplicates based on 'title' and 'company', keeping the first occurrence
df_unique = df_clean.drop_duplicates(subset=['title', 'company'], keep='first')

# Display the header
df_unique.head()
```

```python
# leave only requirements	technologies_expected technologies_optional specializations title	company	 columns
df_final = df_unique[['requirements', 'technologies_expected', 'technologies_optional', 'specializations', 'title', 'company']]
# save it to json 
df_final.to_json('job_offers_cleaned.json', orient='records')
```
